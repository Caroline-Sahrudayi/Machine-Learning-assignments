Our aim to see what boosts the shopping sites success so lets build a model with target variable as amount_spent

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
y=df1['Yearly_Amount_Spent']
x=df1.drop(['Yearly_Amount_Spent'], axis=1)
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=7)
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

## testing our model

```python
y_pred=lr.predict(x_test)
```

```python
plt.scatter(y_test,y_pred)
```

### Evaulating our model

There are different ways to measure the error between the predicted y-values and the actual y-values. We will calculate three kinds of errors using NumPy and Scikit-learn’s metrics:
Mean Absolute Error
Mean Squared Error
Root Mean Squared Error

```python
from sklearn import metrics
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

```python
print(mae,mse,rmse)
```

### Residuals 


#### Although our model seems fairly good at making predictions, we need to make sure everything is okay with our data before the decision-making step.
To do this, let’s plot a histogram of the residuals and make sure it looks normally distributed, using Seaborn’s distplot. The residuals are nothing but the difference between the actual y-values and the predicted y-values.

```python
sns.distplot(y_test-y_pred,bins=50,kde=True,hist=True)
```

#### Residuals are normally distributed means that our error is almost consistent over the entire data set and the predictions hold good at different random levels of the dataset. BAsically means that our dataset is totally random and our models predictions are not 

#### Making the final Decision
